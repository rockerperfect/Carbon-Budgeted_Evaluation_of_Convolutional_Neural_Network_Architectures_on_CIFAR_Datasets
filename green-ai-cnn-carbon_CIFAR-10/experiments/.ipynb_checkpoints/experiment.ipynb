```python
import sys
import os

PROJECT_ROOT = "/content/drive/MyDrive/green-ai-cnn-carbon"
sys.path.append(PROJECT_ROOT)

print("Project root added to Python path")
```

```python
import tensorflow as tf

# Load CIFAR-10
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# One-hot encode labels
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test  = tf.keras.utils.to_categorical(y_test, 10)
```

```python
from models import build_model
from training.train_fixed_epochs import train_fixed_epochs
from training.train_convergence import train_until_convergence
from tracking.carbon_tracker import start_tracker, stop_tracker
from results.logger import log_run

import tensorflow as tf
import numpy as np
import random
```

```python
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

print("Seeds fixed:", SEED)
```

```python
MODELS = ["lenet"]

DATASET_NAME = "cifar10"
TRAINING_MODE = "convergence"   # "fixed" or "convergence"
RUNS = 3

CSV_PATH = f"{PROJECT_ROOT}/results/raw/experiment_runs.csv"
CARBON_LOG_DIR = f"{PROJECT_ROOT}/results/logs/codecarbon"
```

```python
import os
import pandas as pd

# Load existing results if CSV exists
if os.path.exists(CSV_PATH):
    existing_df = pd.read_csv(CSV_PATH)
    print(f"Loaded existing results: {len(existing_df)} rows")
else:
    existing_df = pd.DataFrame()
    print("No existing CSV found. Starting fresh.")
```

```python
for model_name in MODELS:
    print("\n==============================")
    print("Model:", model_name)
    print("==============================")

    for run_id in range(1, RUNS + 1):
        # ----- CHANGE 2: skip existing runs -----
        if not existing_df.empty:
            already_done = (
                (existing_df["model"] == model_name) &
                (existing_df["dataset"] == DATASET_NAME) &
                (existing_df["training_mode"] == TRAINING_MODE) &
                (existing_df["run_id"] == run_id)
            ).any()

            if already_done:
                print(f"Skipping {model_name} | run {run_id} (already exists)")
                continue
        # ---------------------------------------

        print(f"\nRun {run_id}")

        # Change seed per run
        seed = 42 + run_id
        tf.random.set_seed(seed)
        np.random.seed(seed)
        random.seed(seed)

        model = build_model(model_name, num_classes=y_train.shape[1])

        tracker, start_time = start_tracker(
            model_name=model_name,
            dataset_name=DATASET_NAME,
            run_id=run_id,
            output_dir=CARBON_LOG_DIR
        )

        if TRAINING_MODE == "fixed":
            out = train_fixed_epochs(
                model,
                x_train, y_train,
                x_test, y_test,
                epochs=50
            )
        else:
            out = train_until_convergence(
                model,
                x_train, y_train,
                x_test, y_test,
                max_epochs=100
            )

        carbon = stop_tracker(tracker, start_time)

        log_run(
            csv_path=CSV_PATH,
            model=model_name,
            dataset=DATASET_NAME,
            training_mode=TRAINING_MODE,
            run_id=run_id,
            epochs_trained=out["epochs_trained"],
            history=out["history"],
            training_time_sec=out["training_time_sec"],
            carbon_dict=carbon
        )

        print("Logged run successfully")
```
