```python
from models import build_model
from training.train_fixed_epochs import train_fixed_epochs
from tracking.carbon_tracker import start_tracker, stop_tracker
from results.logger import log_run

import tensorflow as tf
import numpy as np
import random
```

```python
BASE_SEED = 42
tf.random.set_seed(BASE_SEED)
np.random.seed(BASE_SEED)
random.seed(BASE_SEED)

print("✅ Base seed fixed:", BASE_SEED)
```

```python
# ===== CIFAR-100 DATASET =====
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar100.load_data()

x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

y_train = tf.keras.utils.to_categorical(y_train, 100)
y_test  = tf.keras.utils.to_categorical(y_test, 100)

print("✅ CIFAR-100 loaded")
print("Train shape:", x_train.shape)
print("Test shape:", x_test.shape)
```

```python
MODEL_NAME = "Model Name"   # CHANGE PER NOTEBOOK
```

```python
DATASET_NAME = "cifar100"
TRAINING_MODE = "fixed"
RUNS = 3
EPOCHS = 50

CSV_PATH = f"{PROJECT_ROOT}/results/raw/experiment_runs.csv"
CARBON_LOG_DIR = f"{PROJECT_ROOT}/results/logs/codecarbon"
```

```python
print(f"🚀 Starting CIFAR-100 training for model: {MODEL_NAME}")

for run_id in range(1, RUNS + 1):
    print(f"\nRun {run_id}")

    seed = BASE_SEED + run_id
    tf.random.set_seed(seed)
    np.random.seed(seed)
    random.seed(seed)

    model = build_model(MODEL_NAME, num_classes=y_train.shape[1])

    tracker, start_time = start_tracker(
        model_name=MODEL_NAME,
        dataset_name=DATASET_NAME,
        run_id=run_id,
        output_dir=CARBON_LOG_DIR
    )

    out = train_fixed_epochs(
        model,
        x_train, y_train,
        x_test, y_test,
        epochs=EPOCHS
    )

    carbon = stop_tracker(tracker, start_time)

    log_run(
        csv_path=CSV_PATH,
        model=MODEL_NAME,
        dataset=DATASET_NAME,
        training_mode=TRAINING_MODE,
        run_id=run_id,
        epochs_trained=out["epochs_trained"],
        history=out["history"],
        training_time_sec=out["training_time_sec"],
        carbon_dict=carbon
    )

    print("✅ Run logged successfully")

print(f"🎉 Completed model: {MODEL_NAME}")print(f"🚀 Starting CIFAR-100 training for model: {MODEL_NAME}")

for run_id in range(1, RUNS + 1):
    print(f"\nRun {run_id}")

    seed = BASE_SEED + run_id
    tf.random.set_seed(seed)
    np.random.seed(seed)
    random.seed(seed)

    model = build_model(MODEL_NAME, num_classes=y_train.shape[1])

    tracker, start_time = start_tracker(
        model_name=MODEL_NAME,
        dataset_name=DATASET_NAME,
        run_id=run_id,
        output_dir=CARBON_LOG_DIR
    )

    out = train_fixed_epochs(
        model,
        x_train, y_train,
        x_test, y_test,
        epochs=EPOCHS
    )

    carbon = stop_tracker(tracker, start_time)

    log_run(
        csv_path=CSV_PATH,
        model=MODEL_NAME,
        dataset=DATASET_NAME,
        training_mode=TRAINING_MODE,
        run_id=run_id,
        epochs_trained=out["epochs_trained"],
        history=out["history"],
        training_time_sec=out["training_time_sec"],
        carbon_dict=carbon
    )

    print("✅ Run logged successfully")

print(f"🎉 Completed model: {MODEL_NAME}")
```
